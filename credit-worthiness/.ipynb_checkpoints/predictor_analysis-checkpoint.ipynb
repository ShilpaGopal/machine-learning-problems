```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/trainingData.csv")
df.head()
```

```python
df.shape
```

```python
pd.DataFrame(df.dtypes).rename(columns={0:'dtype'})
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
# Handling missing data
# primary_business, secondary_business can be set to none

for col in ['primary_business', 'secondary_business']:
    df[col]=df[col].fillna('none')
```

```python
# loan_purpose: there must be reason for the loan, 
# so missing values will be filled with Others
df['loan_purpose'] = df['loan_purpose'].fillna('Others')
```

```python
# droping one negative value
# Prnding
df.groupby('sanitary_availability').size().reset_index(name='counts')
df = df.drop(df[df['sanitary_availability'] == -1.0].index)
df.groupby('sanitary_availability').size().reset_index(name='counts')
```

```python
df.groupby('water_availabity').size().reset_index(name='counts')
```

```python
# sanitary and water availability
pd.crosstab([df.water_availabity], df.sanitary_availability, margins= True).style.background_gradient(cmap='summer_r')
```

```python
df[(df['sanitary_availability'] != 0) & (df['sanitary_availability'] != 1)].isnull().sum()
```

```python
# social_class remove the ., make all upper case,  
# Pending
df['social_class'].replace('.','')
df.groupby('social_class').size().sort_values(ascending=False).sum()
```

```python
# home_ownership and type_of_house
df.groupby('type_of_house').size().sort_values(ascending=False)
```

```python
df.groupby('home_ownership').size().sort_values(ascending=False)
```

```python
#  type_of_house Meaning 
# R - Room
# T1 - Studio apartments
# T2 - One bedroom

pd.crosstab(df.type_of_house,df.home_ownership, margins=True).style.background_gradient(cmap='summer_r')
```

```python
# Data correction for occument counts Extracted from old_dependencies + young_dependencies
df.loc[df.occupants_count >= 100, 'occupants_count'] = df['young_dependents'] + df['old_dependents'] + 1
#sns.countplot(x='occupants_count',hue='type_of_house', data = df)
#df['occupants_count'].value_counts().to_frame().style.background_gradient(cmap='summer')

pd.crosstab(df['occupants_count'], df['type_of_house'],margins=True).style.background_gradient('summer_r')
df.groupby('type_of_house').agg({'occupants_count': ['mean', 'min', 'max']})

plt.subplots(1,1, figsize=(10,10))
sns.boxplot(x="type_of_house", y="occupants_count", data=df)
sns.stripplot(x="type_of_house", y="occupants_count", data=df, jitter=True, edgecolor="gray")

df.loc[df['type_of_house'].isnull(), 'type_of_house'] = np.random.choice(['R', 'T1', 'T2'])
```

```python
# Owership - is there a relationship between monthly expenses and owership 
pd.crosstab(df.home_ownership,df.sanitary_availability, margins=True).style.background_gradient(cmap='summer_r')
```

```python
pd.crosstab(df.home_ownership,df.water_availabity, margins=True).style.background_gradient(cmap='summer_r')

# if they applicant owns house then most likely to have water availability wither full or half

df.loc[(df['water_availabity'].isnull()) & (df['home_ownership'] == 1), 'water_availabity'] = np.random.choice([0.5,1.0])

pd.crosstab(df.home_ownership,df.water_availabity, margins=True).style.background_gradient(cmap='summer_r')
```

```python
# relationship between count of occupant and water and sanity 
pd.crosstab([df.water_availabity, df.sanitary_availability], 
            [df.occupants_count], 
            margins=True).style.background_gradient(cmap='summer_r')
```

```python
pd.crosstab([df.home_ownership], 
            [df.water_availabity, df.sanitary_availability], 
            margins=True).style.background_gradient(cmap='summer_r')
```

```python
# house_ownership is 0 if house area is 0 is wrong 
df.loc[(df['home_ownership'].isnull()) & (df['house_area'] == 0), 'home_ownership'] = 0
#df[(df['home_ownership'] == 1) & (df['house_area'] == 0)]

df.sort_values(by = 'house_area', ascending = False)[:3]

df = df.drop(df[df['Id'] == 37548].index)
df = df.drop(df[df['Id'] == 34764].index)


#plt.subplots(1,1, figsize=(20,20))
#sns.boxplot(x='home_ownership', y='house_area', data = df)
#sns.stripplot(x="home_ownership", y="house_area", data=df, jitter=True, edgecolor="gray")

# Relation ship between Embarked vs Sex, Survived, Pclass
f,ax = plt.subplots(3,2, figsize=(20,15))

sns.countplot('sanitary_availability', data=df, ax=ax[0,0])
ax[0,0].set_title('sanity_availability')

sns.countplot('water_availabity', data=df, ax=ax[0,1])
ax[0,1].set_title('water_availabity')

sns.countplot('sanitary_availability', hue='home_ownership', data=df, ax=ax[1,0])
ax[1,0].set_title('sanitary_availability vs home owership')

sns.countplot('water_availabity', hue='home_ownership', data=df, ax=ax[1,1])
ax[1,1].set_title('water_availabity vs home owership')

sns.countplot('water_availabity', hue='sanitary_availability', data=df, ax=ax[2,0])
plt.show()
```

```python
plt.subplots(figsize=(20, 20))
sns.set(font_scale=1)
sns.heatmap(df.corr(), square= True, cbar= True, annot= True)
plt.show()
```

```python
# If occupant count is 0 then therer is no sanitary 
df.loc[(df['sanitary_availability'].isnull()) & (df['occupants_count'] == 0), 'sanitary_availability'] = 0
df.loc[(df['water_availabity'].isnull()) & (df['occupants_count'] == 0), 'water_availabity'] = 1
```
